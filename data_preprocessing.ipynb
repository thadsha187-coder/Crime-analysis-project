## Loading and Combining Crime Data

The raw data is stored across two folders in the project directory:
- police data 1
- police data 2

Each folder contains monthly CSV files for the selected police forces.All files are automatically loaded and combined into a single dataframe.

After loading the files, the dataset is inspected using:
- shape
- column names
- data types
- summary statistics
- sample records


```python
import os
print(os.getcwd())
print(os.listdir())
```

```python
base_path = "../data/raw"
print("Folder exists:", os.path.exists(base_path))
print("Folders inside raw:", os.listdir(base_path))#check if the folder exists
all_data = [] # creates an empty list 
for data_folder in os.listdir(base_path): #loop through each folder in raw folder 
    data_path = os.path.join(base_path, data_folder)
    if not os.path.isdir(data_path):
        continue
    for month_folder in os.listdir(data_path): #loop through each month folder 
        month_path = os.path.join(data_path, month_folder)
        if not os.path.isdir(month_path): #skips anything that is not a folder 
            continue
        for file in os.listdir(month_path):
            if file.endswith(".csv"):
                file_path = os.path.join(month_path, file)
                print("Loading:", file_path)
                df_temp = pd.read_csv(file_path) # reads csv file
                force_name = file.replace(".csv", "").split("-")[2:-1]
                force_name = " ".join(force_name)
                df_temp["PoliceForce"] = force_name
                all_data.append(df_temp)
df = pd.concat(all_data, ignore_index=True) #combines all the files into one dataframe

print("Data loaded successfully")
print("Rows:", len(df))
print("Columns:", len(df.columns))
```

## Distinguish crime records from outcome records

The combined dataset contains both street crime records and outcome records.  Crime records contain values in 'Crime type', while outcome records primarily contain values in 'Outcome type'.
To preserve both types of information in a single dataframe:
- missing 'Outcome type' values are labelled as 'Outcome not recorded'
- missing 'Crime type' values are labelled as 'Outcome record'
- a new column, 'Record type', is created to classify each row as either 'Crime' or 'Outcome'

```python
df["Crime type"].value_counts(dropna=False)
```

```python
df["Outcome type"].value_counts(dropna=False)
```

```python
df[["Crime type","Outcome type"]].isnull().sum()
```

```python
df.head()
```

```python
df["Outcome type"].fillna("Outcome not recorded", inplace=True)
```

```python
df["Crime type"].fillna("Outcome record", inplace=True)
```

```python
df["LSOA code"].fillna("Unknown", inplace=True)
df["LSOA name"].fillna("Unknown", inplace=True)
```
